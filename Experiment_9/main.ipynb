### <div align="center"> Dataset Description </div>

This experiment uses the Titanic dataset available as an inbuilt dataset in the Seaborn library.

The dataset contains information about passengers who boarded the Titanic ship, including attributes such as age, gender, passenger class, fare, and survival status. It is widely used for data analysis and visualization to identify patterns and relationships within the data.

Source: Seaborn Library (Inbuilt Dataset)  

This dataset is suitable for data visualization as it allows analysis of the distribution of numerical variables across different categories using plots such as box plots.

---

<b> Variable Descriptions: </b>

* survived: Survival status of the passenger (0 = No, 1 = Yes)  
* sex: Gender of the passenger (Categorical)  
* age: Age of the passenger (Numeric)  
* pclass: Passenger class (Numeric/Categorical)  
* fare: Ticket fare paid by the passenger (Numeric)  

---

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

### Data Proprocessing

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(subset=['age'])
```

### Considering only important columns

```python
df = df[['age','sex','survived']]
```

### Final Box Plot

```python
plt.figure(figsize=(7,5))

sns.boxplot(x="sex", y="age", hue="survived", data=df)

plt.title("Age Distribution by Gender and Survival")
plt.show()
```

---

### <div align="center"> Observations and Conclusion </div>

### Observations:

1. Female passengers had a higher survival rate compared to male passengers, as seen from the distribution of survivors in the box plot.
2. The median age of female survivors is slightly higher than that of male survivors, indicating that adult females had better chances of survival.
3. Male passengers show a wider spread in age distribution, especially among those who did not survive, indicating higher variability.
4. Younger passengers (children) are present in both survived and non-survived groups, but survival chances appear relatively better among younger individuals.
5. There are noticeable outliers in both male and female categories, indicating the presence of very young and very old passengers.

### Conclusion:

- In this experiment, a box plot was used to analyze the distribution of age with respect to gender and survival status using the Titanic dataset. The visualization clearly highlights differences in survival patterns between male and female passengers.

- It can be concluded that gender played a significant role in survival, with females having a higher likelihood of survival compared to males. Additionally, age distribution shows that survival was not uniform across all age groups, and certain age ranges had better survival outcomes.

- Overall, box plots proved to be an effective tool for understanding the spread, central tendency, and outliers in the dataset, enabling clear interpretation of patterns in passenger survival.
