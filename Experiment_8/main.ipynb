### <div align="center"> Dataset Description </div>

This experiment uses the Titanic dataset available as an inbuilt dataset in the Seaborn library.

The dataset contains information about passengers who boarded the Titanic ship, including attributes such as age, gender, class, fare, and survival status. It is widely used for data analysis and visualization to identify patterns and relationships within the data.

Source: Seaborn Library (Inbuilt Dataset)  

This dataset is suitable for data visualization as it contains both categorical and numerical variables, allowing the use of plots such as histograms and count plots to understand distributions and patterns.

---

<b> Variable Descriptions: </b>

* survived: Survival status of the passenger (0 = No, 1 = Yes)  
* pclass: Passenger class (1 = First, 2 = Second, 3 = Third)  
* sex: Gender of the passenger (Categorical)  
* age: Age of the passenger (Numeric)  
* sibsp: Number of siblings/spouses aboard (Numeric)  
* parch: Number of parents/children aboard (Numeric)  
* fare: Ticket fare paid by the passenger (Numeric)  
* embarked: Port of embarkation (Categorical)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

### Data Preprocessing

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Data Visulization

```python
survival = df['survived'].value_counts()
ax = sns.barplot(survival)
ax.bar_label(ax.containers[0])
plt.title("Survival Count (Died = 0 and Survived = 1)")
```

### Gender vs Survival

```python
sns.countplot(x="sex",hue="survived",data=df)
plt.title("Survival Based on Gender")
```

### Passenger Class vs Survival

```python
sns.countplot(x='pclass',hue='survived',data=df)
plt.title('Survival Based on Passenger Class')
```

### Age Distribution

```python
sns.histplot(df['age'],bins=30,kde=True)
plt.title('Age Distribution')
```

### Fare Distribution

```python
sns.histplot(df['fare'],bins=30,kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare in $")
plt.ylabel("Frequency")
```
