### <div align="center"> Dataset Description </div>

This experiment uses the Iris dataset obtained from the UCI Machine Learning Repository.

The dataset contains measurements of iris flowers such as sepal length, sepal width, petal length, and petal width for different species. It is widely used for data analysis and visualization to understand feature distributions and identify patterns.

Source: https://archive.ics.uci.edu/ml/datasets/Iris  

This dataset is suitable for data visualization as it contains numerical features that can be analyzed using histograms and box plots to study distributions and detect outliers.

---

<b> Variable Descriptions: </b>

* sepal_length: Length of the sepal (Numeric)  
* sepal_width: Width of the sepal (Numeric)  
* petal_length: Length of the petal (Numeric)  
* petal_width: Width of the petal (Numeric)  
* species: Type of iris flower (Nominal/Categorical)  

---

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris_data.csv')
```

### Data Preprocessing 

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
df.describe()
```

```python
print("Data Types of all columns are : ")
print(df.dtypes)
```

```python
numerical_cols = df.select_dtypes('number').columns
categorical_cols = df.select_dtypes('object').columns
```

### Histogram for each Feature

```python
for col in numerical_cols:
    plt.figure(figsize=(5,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### Box-Plot for each Feature

```python
for col in numerical_cols:
    plt.figure(figsize=(5,4))
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.show()
```

---

### <div align="center"> Observations and Conclusion </div>

### Observations:

1. The features sepal_length and petal_length show a wider distribution compared to sepal_width, indicating greater variability in these measurements.
2. The histogram of petal_length and petal_width shows a clear separation, suggesting that these features are useful for distinguishing between different species.
3. Most features exhibit a slightly skewed distribution rather than a perfectly normal distribution.
4. From the boxplots, sepal_width contains a few outliers, indicating some unusual values in the dataset.
5. Petal_length and petal_width show fewer or no significant outliers and have more consistent distributions.
6. The spread of values is larger in petal-related features compared to sepal-related features.

### Conclusion:

- In this experiment, histograms and boxplots were used to analyze the distribution of numerical features in the Iris dataset. The visualizations helped in understanding the spread, central tendency, and variability of each feature.

- It can be concluded that petal_length and petal_width are more informative features due to their clear distribution patterns and lower presence of outliers. Sepal_width shows some outliers, indicating variability in the data.

- Overall, data visualization techniques such as histograms and boxplots are effective tools for comparing feature distributions and identifying outliers, which are essential steps in exploratory data analysis.
