### <div align="center"> Dataset Description </div>

This experiment uses a sample text document for performing text preprocessing and analysis tasks.

The document contains natural language text which is processed using various Natural Language Processing (NLP) techniques such as tokenization, part-of-speech tagging, stop word removal, stemming, and lemmatization.

This dataset is suitable for text analytics as it allows the application of preprocessing techniques and representation methods such as Term Frequency (TF) and Inverse Document Frequency (IDF).

---

<b> Variable Descriptions: </b>

* Text: Raw input text document (Unstructured Data)  
* Tokens: Individual words obtained after tokenization  
* POS Tags: Grammatical tags assigned to each token  
* Filtered Words: Words after removing stop words  
* Stemmed Words: Words after applying stemming  
* Lemmatized Words: Words after applying lemmatization  
* TF: Term Frequency representing importance of a word in a document  
* IDF: Inverse Document Frequency representing importance of a word across documents

### Import NLTK (Natural Language ToolKit)

```python
import nltk
```

### Download all required resources

### Sample Document

```python
# Sample Document
document = """
Text analytics, also known as text mining, is the process of extracting meaningful insights 
and patterns from unstructured textual data. With the rapid growth of digital content such as 
social media posts, emails, reviews, and articles, analyzing text has become an essential task 
in data science and machine learning.

The process of text analytics begins with preprocessing, which is a crucial step to clean and 
prepare the data. This includes tokenization, where the text is broken down into individual words 
or tokens. Stop word removal is performed to eliminate commonly used words such as 'the', 'is', 
and 'and' that do not contribute much to the meaning of the text.

Further preprocessing involves stemming and lemmatization. Stemming reduces words to their root 
form by removing suffixes, while lemmatization converts words into their meaningful base form 
using linguistic rules. These techniques help in reducing dimensionality and improving the 
efficiency of text analysis.

Another important aspect of text analytics is feature representation. Techniques like Term 
Frequency (TF) measure how frequently a word appears in a document, while Inverse Document 
Frequency (IDF) evaluates how important a word is across multiple documents. The combination 
of TF-IDF helps in identifying the most relevant words in a corpus.

Overall, text analytics plays a significant role in applications such as sentiment analysis, 
spam detection, recommendation systems, and information retrieval. It enables organizations to 
make data-driven decisions by transforming raw text into structured and actionable insights.
"""
```

### Tokenization

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(document)
print("\nTokens:\n", tokens)
```

---

### <div align="center"> POS Tags (Part of Speech Tags) </div>

POS (Part of Speech) tagging assigns grammatical categories to words in a sentence. Below are the important POS tags and their meanings:

<b> Noun Tags: </b>
* NN  : Noun, singular (e.g., "text")
* NNS : Noun, plural (e.g., "analytics")
* NNP : Proper noun, singular (e.g., "India")
* NNPS: Proper noun, plural (e.g., "Americans")

<b> Verb Tags: </b>
* VB  : Verb, base form (e.g., "run")
* VBD : Verb, past tense (e.g., "ran")
* VBG : Verb, gerund/present participle (e.g., "running")
* VBN : Verb, past participle (e.g., "written")
* VBP : Verb, non-3rd person singular present (e.g., "run")
* VBZ : Verb, 3rd person singular present (e.g., "runs")

<b> Adjective Tags: </b>
* JJ  : Adjective (e.g., "meaningful")
* JJR : Adjective, comparative (e.g., "better")
* JJS : Adjective, superlative (e.g., "best")

<b> Adverb Tags: </b>
* RB  : Adverb (e.g., "quickly")
* RBR : Adverb, comparative (e.g., "faster")
* RBS : Adverb, superlative (e.g., "fastest")

<b> Pronoun Tags: </b>
* PRP  : Personal pronoun (e.g., "he", "she", "it")
* PRP$ : Possessive pronoun (e.g., "his", "her")

<b> Determiner Tags: </b>
* DT : Determiner (e.g., "the", "a", "an")

<b> Preposition Tags: </b>
* IN : Preposition or subordinating conjunction (e.g., "in", "on", "of")

<b> Conjunction Tags: </b>
* CC : Coordinating conjunction (e.g., "and", "but", "or")

<b> Miscellaneous Tags: </b>
* CD : Cardinal number (e.g., "one", "100")
* EX : Existential there (e.g., "there is")
* FW : Foreign word
* LS : List item marker
* MD : Modal verb (e.g., "can", "will")
* PDT: Predeterminer
* POS: Possessive ending (e.g., "'s")
* TO : "to"
* UH : Interjection (e.g., "oh", "wow")
* WDT: Wh-determiner (e.g., "which")
* WP : Wh-pronoun (e.g., "who")
* WRB: Wh-adverb (e.g., "where", "when")

<b> Punctuation: </b>
* .  : Sentence-ending punctuation
* ,  : Comma

---

```python
from nltk import pos_tag
pos_tags=pos_tag(tokens)
print("\nPOS Tags:\n", pos_tags)
```

### Stop word removal

```python
from nltk.corpus import stopwords
print("Before Stop Words Removal:\n", len(tokens))
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("After Stop Words Removal:\n", len(filtered_tokens))
```

### Stemming - Finding root word (Fast but Less accurate)

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_tokens]
print("\nStemmed Words:\n", stemmed_words)
```

### Lemmatization - Finding root word (Slow but Accurate)

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("\nLemmatized Words:\n", lemmatized_words)
```

### TF-IDF representation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
```

```python
# Sample corpus (IMPORTANT → must be list of documents)
corpus = [
    "Text analytics is the process of extracting information from text",
    "Text mining involves preprocessing and analysis",
    "Analytics helps in extracting meaningful insights from data"
]
```

```python
tfidf_matrix = vectorizer.fit_transform(corpus)
```

### Creating a DataFrame for clear presentation

```python
import pandas as pd
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
```
