### <div align="center"> Dataset Description </div>

This experiment uses the Iris dataset obtained from Kaggle.

The dataset contains measurements of iris flowers such as sepal length, sepal width, petal length, and petal width for different species. It is widely used for classification problems to predict the species of a flower based on given features.

Source: https://www.kaggle.com/datasets/uciml/iris  

This dataset is suitable for data analytics as it contains numerical features that can be used to implement a Naïve Bayes classification model and evaluate performance using a confusion matrix.

---

<b> Variable Descriptions: </b>

* sepal_length : Length of the sepal (Numeric)  
* sepal_width : Width of the sepal (Numeric)  
* petal_length : Length of the petal (Numeric)  
* petal_width : Width of the petal (Numeric)  
* species : Type of iris flower (Target Variable - Multi-class Classification)

### Dividing Dataset into Independent and Dependent Features

```python
X = df.drop('Species',axis=1)
y = df['Species']
```

### Train-Test Split 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
import numpy as np

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)

print("\nClass-wise TP, FP, FN, TN:")

total = np.sum(cm)

for i in range(len(cm)):
    TP = cm[i, i]
    FP = np.sum(cm[:, i]) - TP
    FN = np.sum(cm[i, :]) - TP
    TN = total - (TP + FP + FN)

    print(f"\nClass {i}:")
    print(f"TP: {TP}")
    print(f"FP: {FP}")
    print(f"FN: {FN}")
    print(f"TN: {TN}")
```

### Performance Metrics

```python
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print("\nPerformance Metrics:")
print(f"Accuracy   : {accuracy:.4f}")
print(f"Error Rate : {error_rate:.4f}")
print(f"Precision  : {precision:.4f}")
print(f"Recall     : {recall:.4f}")

# =========================
# Cross Validation
# =========================
model = GaussianNB()
scores = cross_val_score(model, X, y, cv=5)

print("\nCross Validation:")
print(f"Accuracy for each fold: {scores}")
print(f"Average Accuracy     : {scores.mean():.4f}")
```
