### <div align="center"> Dataset Description </div>

This experiment uses the Social Network Ads dataset obtained from Kaggle.

The dataset contains information about users such as age, estimated salary, and whether they purchased a product or not. It is commonly used for classification problems to predict user behavior based on given features.

<b> Variable Descriptions: </b>
 
* Age : Age of the user (Numeric)  
* EstimatedSalary : Estimated salary of the user (Numeric)  
* Purchased : Whether the user purchased the product (Target Variable - Binary Classification)

### Dividing Dataset into Independent and Dependent Features

```python
X = df.drop('Purchased',axis=1)
y = df['Purchased']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

### Training Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

### Performance Matrix Evaluation

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)

# Extract values
TN, FP, FN, TP = cm.ravel()

print("\nValues from Confusion Matrix:")
print(f"True Negative (TN): {TN}")
print(f"False Positive (FP): {FP}")
print(f"False Negative (FN): {FN}")
print(f"True Positive (TP): {TP}")

# Performance Metrics

# Accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)

# Error Rate
error_rate = 1 - accuracy

# Precision (handle division safely)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0

# Recall
recall = TP / (TP + FN) if (TP + FN) != 0 else 0

print("\nPerformance Metrics:")
print(f"Accuracy      : {accuracy:.4f}")
print(f"Error Rate    : {error_rate:.4f}")
print(f"Precision     : {precision:.4f}")
print(f"Recall        : {recall:.4f}")
```
