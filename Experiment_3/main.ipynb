### 1. Provide summary statistics (mean, median, minimum, maximum, standard deviation) grouped by a categorical variable

```python
df.rename(columns={'Gender':'gender','Age':'age','Education Level':'education','Job Title':'job','Years of Experience':'experience','Salary':'salary'},inplace=True)
```

```python
df.head()
```

### 1.1 : Creating age group using **pd.cut()**

```python
bins = [0,18,30,50,100]
labels = ['teen','young','adult','senior']
df['age_group'] = pd.cut(df['age'],bins=bins,labels=labels)
print(df[['age','age_group']].head())
```

#### 1.2 : Apply **groupby()** function for summary statistics

```python
grouped_by_age = df.groupby('age_group',observed=True)['salary'].agg(['mean', 'median', 'min', 'max', 'std'])
print(grouped_by_age)
```

```python
grouped_by_gender = df.groupby('gender')['salary'].agg(['mean', 'median', 'min', 'max', 'std'])
print(grouped_by_gender)
```

#### 1.3 : Creating list for each category

```python
grouped_by_gender_list = df.groupby('gender')['salary'].apply(list)
grouped_by_age_list = df.groupby('age_group',observed=True)['salary'].apply(list)
print(grouped_by_gender_list)
print(grouped_by_age_list)
```

### 2. Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor’ of iris.csv dataset. 

```python
iris = pd.read_csv('iris_data.csv')
```

```python
iris.head()
```

```python
iris.shape
```

```python
# Group by species
grouped = iris.groupby('Species')
```

#### Grouped Mean

```python
grouped.mean()
```

#### Group Standard Deviation

```python
grouped.std()
```

#### Grouped Percentiles

```python
grouped.quantile([0.25,0.5,0.75])
```
