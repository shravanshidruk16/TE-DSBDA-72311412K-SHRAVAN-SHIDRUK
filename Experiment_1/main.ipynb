### <div align="center"> <b>Dataset Description </b></div>

The dataset used for this experiment is the “Netflix Movies and TV Shows Dataset” obtained from Kaggle.
It contains information about movies and TV shows available on Netflix such as title, director, cast, country, release year, rating, duration, and genre.

Source: https://www.kaggle.com/datasets/shivamb/netflix-shows

This dataset is useful for analyzing content distribution, trends in entertainment, and performing data preprocessing tasks due to the presence of missing values and mixed data types.

<b> Variable Descriptions: </b>
The dataset contains the following variables:

* show_id: Unique ID of the content
* type: Movie or TV Show (Categorical)
* title: Name of the content (Text)
* director: Director name (Categorical/Text)
* cast: Actors involved (Text)
* country: Country of origin (Categorical)
* date_added: Date added to Netflix (Date)
* release_year: Year of release (Numeric)
* rating: Content rating (Categorical)
* duration: Duration of content (Text)
* listed_in: Genre (Categorical)
* description: Summary of the content (Text)

### 4. Type Conversion

```python
df["date_added"] = df["date_added"].str.strip()     # 1. removing extra spaces from dates
df["date_added"] = pd.to_datetime(df["date_added"]) # 2. converting date from obj to datetime data type
df["release_year"] = df["release_year"].astype(int) # 3. converting year into int
```

### 5. Data Normalization

```python
import warnings
warnings.filterwarnings('ignore')


# Extract numeric duration
df['duration'] = df['duration'].astype(str).str.extract('(\d+)')

# Convert to numeric
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# Fill missing values
df['duration'] = df['duration'].fillna(df['duration'].mean())

# Normalize
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df[['duration', 'release_year']] = scaler.fit_transform(
    df[['duration', 'release_year']]
)
```

### 6. Converting Categorical to Numerical features

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['type'] = le.fit_transform(df['type'])
df['rating'] = le.fit_transform(df['rating'])
df['country'] = le.fit_transform(df['country'].astype(str))
```
